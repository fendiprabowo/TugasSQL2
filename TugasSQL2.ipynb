Load sample database

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## One to many relationship 

### 1. Report the account representative for each customer. 

```python
run("""
SELECT 
    customerNumber, 
    customerName, 
    CONCAT(firstName, ' ', lastName) AS representativeName,
    employeeNumber AS representativeNumber
FROM
    customers c
        INNER JOIN
    employees e ON c.salesRepEmployeeNumber = e.employeeNumber
""")
```

### 2. Report total payments for Atelier graphique.

```python
run("""
SELECT 
    customerNumber, 
    customerName, 
    SUM(amount) AS totalPayments
FROM
    customers
        INNER JOIN
    payments USING(customerNumber)
WHERE customerName='Atelier graphique'
""")
```

### 3. Report the total payments by date 

```python
run("""
SELECT 
    customerNumber, 
    customerName,
    paymentDate,
    SUM(amount) AS totalPayments
FROM
    customers
        INNER JOIN
    payments USING(customerNumber)
GROUP BY
    paymentDate
""")
```

### 4. Report the products that have not been sold.

```python
run("""
SELECT 
    productCode, 
    productName,
    orderNumber AS statusSelling
FROM
    products
        LEFT JOIN
    orderdetails USING(productCode)
WHERE
    orderNumber IS NULL
""")
```

### 5. List the amount paid by each customer. 

```python
run("""
SELECT 
    customerNumber, 
    customerName,
    SUM(amount) AS totalPayments
FROM
    customers
        INNER JOIN
    payments USING(customerNumber)
GROUP BY
    customerNumber
""")
```

### 6. How many orders have been placed by Herkku Gifts? 

```python
run("""
SELECT 
    customerNumber, 
    customerName, 
    COUNT(orderNumber) AS totalOrders
FROM
    customers
        INNER JOIN
    orders USING(customerNumber)
WHERE customerName='Herkku Gifts'
""")
```

### 7. Who are the employees in Boston? 

```python
run("""
SELECT 
    employeeNumber, 
    CONCAT(firstName, ' ', lastName) AS employeeName,
    email,
    officeCode,
    city,
    addressLine1
FROM
    employees
        INNER JOIN
    offices USING(officeCode)
WHERE city='Boston'
""")
```

### 8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first. 

```python
run("""
SELECT 
    customerNumber, 
    customerName, 
    SUM(amount) AS totalPayments
FROM
    customers
        INNER JOIN
    payments USING(customerNumber)
GROUP BY
    customerNumber
HAVING 
    totalPayments>100000
ORDER BY
    totalPayments DESC
""")
```

### 9. List the value of 'On Hold' orders. 

```python
run("""
SELECT 
    orderNumber, 
    orderDate, 
    status,
    SUM(quantityOrdered*priceEach) AS totalValue
FROM
    orders
        INNER JOIN
    orderdetails USING(orderNumber)
WHERE
    status='On Hold'
GROUP BY
    orderNumber
""")
```

### 10. Report the number of orders 'On Hold' for each customer. 

```python
run("""
SELECT 
    c.customerNumber, 
    customerName,
    COUNT(orderNumber) AS totalOnHold
FROM
    customers c
        LEFT JOIN
    orders o ON c.customerNumber=o.customerNumber
    AND o.status='On Hold'
GROUP BY
    customerNumber
ORDER BY
    totalOnHold DESC
""")
```

## Many to many relationship 

### 1. List products sold by order date. 

```python
run("""
SELECT 
    pd.productCode, 
    productName,
    o.orderNumber,
    orderDate,
    status
FROM
    products pd
        INNER JOIN
    orderdetails od ON pd.productCode=od.productCode
        INNER JOIN
    orders o ON od.orderNumber=o.orderNumber
        AND o.status='Shipped'
ORDER BY
    orderDate DESC
""")
```

### 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck. 

```python
run("""
SELECT 
    o.orderNumber, 
    orderDate,
    status,
    productName
FROM
    orders o
        INNER JOIN
    orderdetails od ON od.orderNumber=o.orderNumber
        INNER JOIN
    products pd ON pd.productCode=od.productCode
        AND pd.productName='1940 Ford Pickup Truck'
ORDER BY
    orderDate DESC
""")
```

### 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT 
    c.customerNumber, 
    customerName,
    o.orderNumber,
    orderDate,
    SUM(quantityOrdered*priceEach) AS totalValue
FROM
    customers c
        INNER JOIN
    orders o ON c.customerNumber=o.customerNumber
        INNER JOIN
    orderdetails od ON od.orderNumber=o.orderNumber
GROUP BY
    orderNumber
HAVING
    totalValue>25000
ORDER BY
    orderDate DESC
""")
```

### 4. Are there any products that appear on all orders? 

```python
run("""
SELECT 
    productCode,
    productName,
    COUNT(productCode) AS totalAppear
FROM 
    orderdetails 
        INNER JOIN
    products USING(productCode)
GROUP BY 
    productCode 
HAVING 
    totalAppear=(SELECT COUNT(*) from orders);
""")
```

### 5. List the names of products sold at less than 80% of the MSRP. 

```python
run("""
SELECT 
    pd.productCode, 
    productName,
    MSRP,
    o.orderNumber,
    orderDate,
    status,
    priceEach
FROM
    products pd
        INNER JOIN
    orderdetails od ON pd.productCode=od.productCode
        INNER JOIN
    orders o ON od.orderNumber=o.orderNumber
        AND o.status='Shipped'
WHERE
    priceEach < (MSRP * .8)
ORDER BY
    orderDate DESC
""")
```

### 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice) 

```python
run("""
SELECT 
    pd.productCode, 
    productName,
    buyPrice,
    o.orderNumber,
    orderDate,
    status,
    priceEach
FROM
    products pd
        INNER JOIN
    orderdetails od ON pd.productCode=od.productCode
        INNER JOIN
    orders o ON od.orderNumber=o.orderNumber
        AND o.status='Shipped'
WHERE
    priceEach > (buyPrice * 2)
ORDER BY
    orderDate DESC
""")
```

### 7. List the products ordered on a Monday. 

```python
run("""
SELECT 
    pd.productCode, 
    productName,
    o.orderNumber,
    orderDate,
    DAYNAME(orderDate) AS dayOrder,
    status
FROM
    products pd
        INNER JOIN
    orderdetails od ON pd.productCode=od.productCode
        INNER JOIN
    orders o ON od.orderNumber=o.orderNumber
WHERE
    DAYNAME(orderDate)='Monday'
ORDER BY
    orderDate DESC
""")
```

### 8. What is the quantity on hand for products listed on 'On Hold' orders? 

```python
run("""
SELECT 
    orderNumber, 
    orderDate, 
    status,
    SUM(quantityOrdered) AS totalQuantity
FROM
    orders
        INNER JOIN
    orderdetails USING(orderNumber)
WHERE
    status='On Hold'
GROUP BY
    orderNumber
""")
```
